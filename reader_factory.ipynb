```python
# 1. Base Class (Blueprint)
class DataSource:
    """
    Abstract class for all data sources
    """
    def __init__(self, path):
        self.path = path

    def get_data_frame(self):
        """
        Abstract method, to be implemented by child classes
        """
        raise ValueError("Not Implemented")

# 2. Child Classes (Specific Formats)
class CSVDataSource(DataSource):
    def get_data_frame(self):
        return (
            spark.read
            .format("csv")
            .option("header", "True")
            .load(self.path)
        )

class ParquetDataSource(DataSource):
    def get_data_frame(self):
        return (
            spark.read
            .format("parquet")
            .load(self.path)
        )

class ORCDataSource(DataSource):
    def get_data_frame(self):
        return (
            spark.read
            .format("orc")  # Corrected format to "orc"
            .load(self.path)
        )

class DeltaDataSource(DataSource):
    def get_data_frame(self):
        return (
            spark.read
            .format("delta")
            .load(self.path)
        )

# 3. Factory Method (The Decision Maker)
def get_data_source(data_type, file_path):
    if data_type == "csv":
        return CSVDataSource(file_path)
    elif data_type == "parquet":
        return ParquetDataSource(file_path)
    elif data_type == "orc":
        return ORCDataSource(file_path)
    elif data_type == "delta":
        return DeltaDataSource(file_path)
    else:
        raise ValueError(f"Data type {data_type} not supported")
```
