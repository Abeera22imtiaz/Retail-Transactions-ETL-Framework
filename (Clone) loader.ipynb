```python
class AbstractLoader:
    def __init__(self, transformedDF):
        self.transformedDF = transformedDF

    def sink(self):
        pass

class AirPodsAfterIphoneLoader(AbstractLoader):
    
    def sink(self):
        
        get_sink_source(
            sink_type='dbfs',
            df=self.transformedDF,
            path="/Volumes/workspace/default/data/output/airpodsAfterIphone",
            method='overwrite'
        ).load_data_frame()
        
class OnlyAirpodsAndIphoneLoader(AbstractLoader):
    
    def sink(self):

        params={
            'PartitionByColumns':['location']
        }
        
        get_sink_source(
            sink_type='dbfs_partition',
            df=self.transformedDF,
            path="/Volumes/workspace/default/data/output/OnlyAirpodsAndIphone",
            method='overwrite',
            params=params
        ).load_data_frame()

        get_sink_source(
            sink_type = "delta",
            df = self.transformedDF,
            path = "default.onlyAirPodsAndIphone",
            method = "overwrite",
            ).load_data_frame()
        
        
```
